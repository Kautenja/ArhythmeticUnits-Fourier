# FFT

The "Fast" Fourier Transform (FFT.) is an $O(N \log_2{N})$ re-formulation of the DFT, which has an $O(N^2)$ complexity. The FFT enables real-time applications that would otherwise be impossible to achieve with the DFT. Furthermore, many implementations of the FFT can be executed in-place, which has direct implications on real-world performance.

```python
import numpy as np
import scipy.fftpack as fft
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sine(frequency: float, sample_rate: float, num_samples: int) -> np.ndarray:
    """
    Generate a sinusoidal waveform.

    Args:
        frequency: The fundamental frequency of the waveform.
        sample_rate: The sample rate to generate the waveform at.
        num_samples: The number of samples of the waveform to generate.

    Returns:
        A sinusoid of length $N$ with frequency $f Hz$ sampled at $f_s Hz$.

    """
    x = np.arange(num_samples)
    return np.cos(2 * np.pi * frequency * x / sample_rate)
```

```python
FREQUENCY = 441      # The fundamental frequency
SAMPLE_RATE = 44100  # The sampling rate
NUM_SAMPLES = 4096   # The number of samples to yield
SINUSOID = sine(FREQUENCY, SAMPLE_RATE, NUM_SAMPLES)
```

```python
plt.figure(figsize=(15, 2))
plt.plot(SINUSOID)
plt.xlabel('n')
plt.ylabel('x[n]')
plt.title(f'Sinusoid with fundamental frequency of {FREQUENCY:d}Hz sampled at {SAMPLE_RATE:d}Hz')
plt.tight_layout()
```
